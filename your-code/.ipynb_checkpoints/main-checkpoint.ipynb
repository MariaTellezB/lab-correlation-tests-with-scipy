# Challenge 1 - Analysis of Variance

In this part of the lesson, we will perform an analysis of variance to determine whether the factors in our model create a significant difference in the group means. We will be examining a dataset of FIFA players. We'll start by loading the data using the code in the cell below.

```python
# Run this code:

fifa = pd.read_csv('fifa.csv')
```

Let's examine the dataset by looking at the `head`.

```python
# Your code here:

fifa.head()
```

```python
fifa.tail()
```

Player's values are expressed in euros. We would like this column to be numeric. Therefore, let's create a numeric value column. Do this by stripping all non-numeric characters from each cell and assign this new data to `ValueNumeric`. There is no need to multiply the value to be expressed in millions or thousands but converting them carefully into the same scale. 

```python
# Your code here:

fifa.Value.unique()
```

```python
fifa['ValueNumeric'] = fifa.Value.str.replace('[€K]','')
```

```python
fifa.tail()
```

```python
fifa['ValueNumeric'] = np.where(fifa['ValueNumeric'].str.find('M') !=-1, fifa['ValueNumeric'].str.replace('M','').astype('float64'), 
                            fifa['ValueNumeric'].str.replace('M','').astype('float64')/1000)
```

```python
fifa.head()
```

```python
fifa.tail()
```

#### We'd like to determine whether a player's preffered foot and position have an impact on their value. 

Using the `statsmodels` library, we are able to produce an ANOVA table without munging our data. Create an ANOVA table with value as a function of position and preferred foot. Recall that pivoting is performed by the `C` function.

Hint: For columns that have a space in their name, it is best to refer to the column using the dataframe (For example: for column `A`, we will use `df['A']`).

```python
# Your code here:

import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
```

ANOVA Hypotheses

Ho = Groups means are equal 
Hi = at least, one group mean is different from other groups

F Tests is a test for comparing variances.

```python
model = ols('ValueNumeric ~ C(Position)*C(fifa["Preferred Foot"])', data = fifa).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

What is your conclusion from this ANOVA?

#### Your conclusions here:

The p-values obtained from the ANOVA analysis, denote that the test is statistically significant, the numeric values differ by position and preferred foot. 

After looking at a model of both preffered foot and position, we decide to create an ANOVA table for nationality. Create an ANOVA table for numeric value as a function of nationality.

```python
# Your code here:

model = ols('ValueNumeric ~ C(Nationality)', data = fifa).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

What is your conclusion from this ANOVA?

Again, the p-value indicates that the tests is statisticaly significant, and that the numeric values differ by nationality. 

# Challenge 2 - Linear Regression

Our goal with using linear regression is to create a mathematical model that will enable us to predict the outcome of one variable using one or more additional independent variables.

We'll start by ensuring there are no missing values. Examine all variables for all missing values. If there are missing values in a row, remove the entire row.

```python
# Your code here:

fifa.shape
```

Using the FIFA dataset, in the cell below, create a linear model predicting value using stamina and sprint speed. create the model using `statsmodels`. Print the model summary.

Hint: remember to add an intercept to the model using the `add_constant` function.

```python
# Your code here:

X = sm.add_constant(fifa[['Stamina','SprintSpeed']])
Y = fifa.ValueNumeric

model = sm.OLS(Y,X).fit()

print(model.summary())
```

```python
'''The F value in regression is the result of a test where the null hypothesis is that all of the regression coefficients are equal to zero. 
In other words, the model has no predictive capability. Basically, the f-test compares your model with zero predictor variables (the intercept only model), 
and decides whether your added coefficients improved the model. If you get a significant result, then whatever coefficients you included in your model improved the model’s fit.

Read your p-value first. If the p-value is small (less than your alpha level), you can reject the null hypothesis. 
Only then should you consider the f-value. If you don’t reject the null, ignore the f-value.'''
```

Report your findings from the model summary. In particular, report about the model as a whole using the F-test and how much variation is predicted by the model using the r squared.

#### Your conclusions here:

By the value of the R-squared, it can be said that is not the best model, it can only describes about the 4.8% of variation in the data.

Being the P-value euqal to 0, the null hypothesis is rejected, and with the F-value, combined with R^2, it can be stated that only one of the variables (stamina or sprintspeed) is 
statisticaly significant. 

Next, create a second regression model predicting value using potential. Create the model using `statsmodels` and print the model summary. Remember to add a constant term.

```python
# Your code here:


X = sm.add_constant(fifa.Potential)
Y = fifa.ValueNumeric

model = sm.OLS(Y,X).fit()

print(model.summary())
```

statisticaly significant.
